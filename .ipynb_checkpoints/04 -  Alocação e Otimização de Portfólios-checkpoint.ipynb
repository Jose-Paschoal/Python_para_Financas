```python
## definir aleatoriamente quanto de dinheiro será investido em cada empresa.

## no final faz uma normalização dos pesos gerados para que a soma não ultrapasse 1.

## normalizando o dataset conforme já fizemos anteriormente

def alocacao_ativos(df, dinheiro_total):
    df = df.copy()
    
    pesos = np.random.random(len(df.columns)-1)
    pesos = pesos/pesos.sum()
    
    print(pesos, pesos.sum())
    
    colunas = df.columns[1:]
    for i in colunas:
        df[i] = (df[i]/df[i][0])
        
    for i, acao in enumerate(df.columns[1:]):
        df[acao] = df[acao] * pesos[i] * dinheiro_total
        
    df['soma valor'] = df.sum(axis=1)
    
    return df
```

```python
alocacao_ativos(pd.read_csv('acoes.csv'), 5000)
```
