# Base de dados expandida 

```python
acoes = ['GOLL4.SA', 'CVCB3.SA', 'WEGE3.SA', 'MGLU3.SA', 'TOTS3.SA', 'BOVA11.SA']
```

```python
acoes_df = pd.DataFrame()
for acao in acoes:
    acoes_df[acao] = yf.download(acao, start='2015-01-01')['Close']
```

```python
acoes_df
```

## Renomeando as ações do DataFrame

```python
acoes_df = acoes_df.rename(columns={'GOLL4.SA': 'GOL', 'CVCB3.SA' : 'CVC', 'WEGE3.SA' : 'WEG', 
                                    'MGLU3.SA' : 'MGLU', 'TOTS3.SA' : 'TOTS', 'BOVA11.SA': 'BOVA'})
```

```python
acoes_df
```

```python
#checando se há valores nulos

acoes_df.isnull().sum()
```

```python
acoes_df.to_csv('acoes.csv')
```

```python
# estatísticas do conjunto de ações

acoes_df.describe()
```

```python
#importando o arquivo

acoes_df = pd.read_csv('acoes.csv')
```

# Visualização dos dados por meios gráficos

```python
#histograma
sns.histplot(acoes_df['GOL']);
```

```python
#definindo um for para automatizar a geração de gráficos de um dataframe

# kde inserido abaixo gera uma linha de densidade no gráfico

plt.figure(figsize=(10,50))
i = 1
for i in np.arange(1, len(acoes_df.columns)):
    plt.subplot(7, 1, i + 1)
    sns.histplot(acoes_df[acoes_df.columns[i]], kde = True)
    plt.title(acoes_df.columns[i])
```

```python
#Boxplot - conseguimos enxergar os quartis e mediana nesse gráfico

sns.boxplot(x = acoes_df['GOL']);
```

```python
acoes_df['GOL'].describe()
```

```python
plt.figure(figsize=(10,50))
i = 1
for i in np.arange(1, len(acoes_df.columns)):
    plt.subplot(7, 1, i + 1)
    sns.boxplot(x = acoes_df[acoes_df.columns[i]])
    plt.title(acoes_df.columns[i])
```

```python
#Gráfico de linha

acoes_df.plot(x = 'Date', figsize = (15, 7), title = 'Histórico do Preço das Ações');
```

## Normalização dos dados para melhor comparação entre os ativos

```python
#objetivo de analisar as ações através do mesmo ponto de partida e com isso identificar as melhores oportunidades de valorização

# começando na coluna 1 já que a coluna 0 no dataframe são as datas.

acoes_df_normalizado = acoes_df.copy()
for i in acoes_df_normalizado.columns[1:]:
    acoes_df_normalizado[i] = acoes_df_normalizado[i] / acoes_df_normalizado[i][0]
```

```python
acoes_df_normalizado
```

```python
acoes_df_normalizado.plot(x = 'Date', figsize = (15, 7), title = 'Histórico do Preço das Ações - normalizado');
```
